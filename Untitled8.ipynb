```python
# 1. Setup and Library Import
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import streamlit as st
```

```python
# 2. Data Loading
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# 3. Data Merging
credits.rename(columns={'movie_id': 'id'}, inplace=True)
merged_data = pd.merge(movies, credits, on='id')
merged_data
```

```python
# 4. Initial Data Exploration
print(merged_data.head())
print(merged_data['crew'].head())
```

```python
# 5. Column Selection
selected_columns = merged_data[['genres', 'id', 'keywords', 'original_title', 'overview', 'cast', 'crew']]
selected_columns = selected_columns.rename(columns={'original_title': 'movie_title'})

```

```python
# 6. Data Cleaning
selected_columns = selected_columns.copy()
selected_columns.dropna(inplace = True)
selected_columns.drop_duplicates(inplace = True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'selected_columns' contains your movie data

# Flatten the 'genres' column and count genre occurrences
all_genres = [genre for genres_list in selected_columns['genres'] for genre in genres_list]
genre_counts = pd.Series(all_genres).value_counts()

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Distribution of Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 7. Feature Engineering
# Genres Extraction
def extract_genres(genres):
    return [genre['name'] for genre in ast.literal_eval(genres)]

selected_columns['genres'] = selected_columns['genres'].apply(extract_genres)
```

```python
# Keywords Extraction
def extract_keywords(keywords):
    return [keyword['name'] for keyword in ast.literal_eval(keywords)]

selected_columns['keywords'] = selected_columns['keywords'].apply(extract_keywords)
```

```python
# Cast Extraction
def extract_cast(cast):
    return [member['name'] for member in ast.literal_eval(cast)]

selected_columns['cast'] = selected_columns['cast'].apply(extract_cast)
```

```python
def extract_crew(crew):
       try:
           return [member['name'] for member in ast.literal_eval(crew)]
       except (SyntaxError, ValueError):
           return []

selected_columns['crew'] = selected_columns['crew'].apply(extract_crew)
```

```python
# 8. Preprocessing for Recommendation
# Combine textual features
selected_columns['combined_features'] = selected_columns['genres'].apply(lambda x: ' '.join(x)) + ' ' + \
                                         selected_columns['overview'] + ' ' + \
                                         selected_columns['keywords'].apply(lambda x: ' '.join(x))
```

```python
# Vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(selected_columns['combined_features'])

```

```python
# Cosine Similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
# 9. Model Development
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = selected_columns[selected_columns['movie_title'] == title].index[0]  # Use the correct column name
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]  # Get top 5 recommendations
    movie_indices = [i[0] for i in sim_scores]
    return selected_columns['movie_title'].iloc[movie_indices]
```

```python
import streamlit as st
# ... (your existing code) ...

# 11. Deployment Using Streamlit
st.title('Movie Recommender System')

user_input = st.text_input("Enter a movie title:")

if user_input:
    try:
        recommendations = get_recommendations(user_input)
        st.write("Recommendations for", user_input + ":")  # Display input movie
        for movie in recommendations:
            st.write(movie)
    except IndexError:
        st.write("Movie not found in the database. Please try another title.")

# ... (rest of your code) ...
```
